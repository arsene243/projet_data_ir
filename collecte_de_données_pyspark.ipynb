```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import count, desc, col, max, struct
import matplotlib.pyplot as plts
```

Création d'une session spark

```python
spark = SparkSession.builder.appName("spark_app").getOrCreate()
```

Importation du jeu de données

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
weatherHistory_csv_path = "/content/drive/MyDrive/projet_data_ir/weatherHistory.csv"
weatherHistory_df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load(weatherHistory_csv_path)
```

```python
weatherHistory_df.show()
```

```python
weatherHistory_df.count()
```

# données manquantes


```python
from pyspark.sql.functions import col, sum, when

# Compter les valeurs nulles par colonne
weatherHistory_df.select([sum(when(col(c).isNull(), 1).otherwise(0)).alias(c) for c in weatherHistory_df.columns]).show()
```
